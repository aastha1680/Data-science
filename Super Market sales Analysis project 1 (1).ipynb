```python

print(df.columns)


df['Year_of_Birth'] = pd.to_datetime(df['Time']).dt.year

# Calculating age by subtracting the birth year from the current year
df['Age'] = pd.datetime.now().year - df['Year_of_Birth']


# Plotting age distribution
plt.figure(figsize=(12, 6))
sns.histplot(df['Age'], bins=20, kde=True)
plt.title('Age Distribution of Customers')
plt.xlabel('Age')
plt.ylabel('Frequency')
plt.show()

# Correlation between age and gross income
sns.scatterplot(x=df['Age'], y=df['gross income'])
plt.title('Correlation between Age and Gross Income')
plt.xlabel('Age')
plt.ylabel('Gross Income')
plt.show()

```

```python

print(df.columns)


plt.figure(figsize=(8, 5))
sns.countplot(x=df['Customer type'])  
plt.title('Distribution of Customers based on Member Type')
plt.xlabel('Member Type')
plt.ylabel('Count')
plt.show()

# Analyzing the impact of member type on gross income
sns.boxplot(x=df['Customer type'], y=df['gross income'])
plt.title('Customer Type vs Gross Income')
plt.xlabel('Customer Type')
plt.ylabel('Gross Income')
plt.show()
```

```python

print(df.columns)

plt.figure(figsize=(14, 8))
sns.countplot(x='Product line', data=df, palette='viridis')
plt.title('Top-Selling Products within Each Product Line')
plt.xlabel('Product Line')
plt.ylabel('Count')
plt.xticks(rotation=45)
plt.legend(title='Product', bbox_to_anchor=(1.05, 1), loc='upper left')
plt.show()



# Analyze profitability of each product line
plt.figure(figsize=(14, 8))
sns.barplot(x='Product line', y='gross income', data=df, ci=None, estimator=sum, palette='muted')
plt.title('Profitability of Each Product Line')
plt.xlabel('Product Line')
plt.ylabel('Total Gross Income')
plt.xticks(rotation=45)
plt.show()
```

```python
# Relationship between ratings and time of day
plt.figure(figsize=(12, 6))
sns.boxplot(x='Hour', y='Rating', data=df)
plt.title('Relationship between Ratings and Time of Day')
plt.xlabel('Hour of the Day')
plt.ylabel('Customer Rating')
plt.show()

# Relationship between ratings and day of the week
plt.figure(figsize=(12, 6))
sns.boxplot(x='weekday', y='Rating', data=df, order=['Monday', 'Tuesday', 'Wednesday', 'Thursday', 'Friday', 'Saturday', 'Sunday'])
plt.title('Relationship between Ratings and Day of the Week')
plt.xlabel('Day of the Week')
plt.ylabel('Customer Rating')
plt.show()

# Relationship between ratings and branch
plt.figure(figsize=(8, 6))
sns.boxplot(x='Branch', y='Rating', data=df)
plt.title('Relationship between Ratings and Branch')
plt.xlabel('Branch')
plt.ylabel('Customer Rating')
plt.show()

# Relationship between ratings and method of payment
plt.figure(figsize=(10, 6))
sns.boxplot(x='Payment', y='Rating', data=df)
plt.title('Relationship between Ratings and Method of Payment')
plt.xlabel('Method of Payment')
plt.ylabel('Customer Rating')
plt.show()

# Relationship between ratings and customer type
plt.figure(figsize=(10, 6))
sns.boxplot(x='Customer type', y='Rating', data=df)
plt.title('Relationship between Ratings and Customer Type')
plt.xlabel('Customer Type')
plt.ylabel('Customer Rating')
plt.show()
```
